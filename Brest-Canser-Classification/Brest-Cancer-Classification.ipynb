#### Importing the Dependencies module

```python
import numpy as np
import pandas as pd
import matplotlib as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

#### Data Collection and Processing

```python
#load data in dataframe
df = pd.read_csv("data.csv")
df.head()
```

```python
df['diagnosis'].value_counts()
```

```python
df['diagnosis'] = df['diagnosis'].map({'B': 1, 'M': 0})
df = df.drop('Unnamed: 32', axis=1)
df = df.drop('id', axis=1)
```

```python
df.head()
```

### seprating the featuresand target

```python
x = df.drop('diagnosis',axis=1)
y = df['diagnosis']
```

```python
print(x)
```

```python
print(y)
```

#### splitting data into traing data and testing data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

## Model training


#### Logistic regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model using traing data

model.fit(x_train,y_train)
```

### Model evaluation

#### Accuracy score

```python
# accuracy on traing data

x_train_pridiction = model.predict(x_train)

acc_train = accuracy_score(y_train,x_train_pridiction)
acc_train
```

```python
# accuracy on testing data
x_test_pridiction = model.predict(x_test)
acc_test = accuracy_score(x_test_pridiction,y_test)
acc_test
```

### Building a Pridictive System

```python
input_data = (12.05,14.63,78.04,449.3,0.1031,0.09092,0.06592,0.02749,0.1675,0.06043,0.2636,0.7294,1.848,19.87,0.005488,0.01427,0.02322,0.00566,0.01428,0.002422,13.76,20.7,89.88,582.6,0.1494,0.2156,0.305,0.06548,0.2747,0.08301
)
# change the input data to a numpy arry
input_data_as_numpy_arry = np.array(input_data)

#reshape the numpy arry as we are pridicting for one datapoint
input_data_reshaped = input_data_as_numpy_arry.reshape(1,-1)

# predict the output
prediction = model.predict(input_data_reshaped)
print(prediction)

if prediction[0] == 0:
    print("Breast cancer is Malignant")
else:
    print("Breast cancer is Benign")
```
